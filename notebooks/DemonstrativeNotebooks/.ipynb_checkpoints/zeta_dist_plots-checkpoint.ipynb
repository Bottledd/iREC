```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zipf
import matplotlib
```

```python
%cd ../..
```

```python
%matplotlib inline
```

```python
plt.rcParams.update({'font.weight': 'normal'})
plt.rcParams.update({'font.size': 16})
plt.rcParams.update({'lines.linewidth' : 2})
```

```python
f, ax = plt.subplots(1, figsize=(8,6))
ss = [2, 3, 4, 5]
xs = np.arange(1, 11)
for s in ss:
    dist = zipf(s)
    
    ax.plot(xs, dist.pmf(xs), 'd-')
ax.set_yscale("log")
ax.set_xscale("log")
ax.set_xticks(xs)
ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax.set_xlabel('k')
ax.set_ylabel('Probability')
ax.legend([r'$s=2$', r'$s=3$', r'$s=4$', r'$s=5$'])
f.tight_layout()
f.savefig("Figures/Thesis/Zeta_Dist.pdf", bbox_inches='tight')
```
