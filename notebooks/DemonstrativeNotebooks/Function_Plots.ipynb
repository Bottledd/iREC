```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import torch
import numpy as np
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 16}

plt.rcParams.update({'font.family': 'normal', 'font.weight' : 'normal', 'font.size' : 16})
```

```python
rhos = np.linspace(-6, 6, 1000)
```

```python
def softplus_np(x): return np.log1p(np.exp(-np.abs(x))) + np.maximum(x, 0)
```

```python
f, ax = plt.subplots(1, figsize=(8,6))
fs = 16
ax.plot(rhos, softplus_np(rhos))
ax.set_yticks(np.arange(-6, 7, 2))
ax.set_ylabel('Output', fontsize=fs)
ax.set_xlabel('Input', fontsize=fs)
ax.set_title('Softplus Function', fontsize=fs)
ax.grid()
f.tight_layout()
```
